```python
import psycopg2
from psycopg2 import sql

# PostgreSQL connection details
host = "localhost"
database = "employee"
user = "postgres"
password = "7"
table_name = 'youtubevide'
csv_file = 'city_chart.csv'
# Establish a connection to the PostgreSQL database
conn = psycopg2.connect(host=host, database=database, user=user, password=password)

# Create a cursor object to interact with the database
cursor = conn.cursor()

# Create the table if it doesn't exist
create_table_sql = """
CREATE TABLE IF NOT EXISTS {} (
    date DATE,
    city_id VARCHAR(255),
    city_name VARCHAR(255),
    views INTEGER
);
""".format(table_name)
cursor.execute(create_table_sql)

# Truncate the table if needed
cursor.execute(sql.SQL("TRUNCATE {} RESTART IDENTITY;").format(sql.Identifier(table_name)))

# Prepare the INSERT statement
insert_sql = sql.SQL("INSERT INTO {} (date, city_id, city_name, views) VALUES (%s, %s, %s, %s);").format(sql.Identifier(table_name))

# Read the CSV file and insert the data row by row
with open(csv_file, 'r') as file:
    next(file)  # Skip the header row
    for line in file:
        data = line.strip().split(',')
        cursor.execute(insert_sql, data)

# Commit the changes
conn.commit()

# Close the cursor and database connection
cursor.close()
conn.close()

print("CSV file imported successfully into the PostgreSQL table.")
```
